```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

### Check the data types of the columns. 

```python
data = pd.read_csv('marketing_customer_analysis.csv', sep=",")
data.head()
```

```python
data=data.drop(['Unnamed: 0'], axis=1)
data.head()
```

 ### X-y split.

```python
y = data['Total Claim Amount']
X = data.drop(['Total Claim Amount'], axis=1)
X.head()
```

```python
X_num = X.select_dtypes(include = np.number)
X_cat = X.select_dtypes(include = 'object')
```

```python
X_cat.head()
```

```python
#X_cat.drop(['Customer', 'State'], axis=1)
X_cat=X_cat.drop(columns=['Customer', 'State','Effective To Date'])
X_cat
```

### One Hot/Label Encoding (categorical).

```python
X_cat= X_cat.replace(np.nan, 'Unknown')
```

```python
#data['Gender'].unique()
for column in X_cat.columns:
    print(X_cat[column].unique())
```

```python
#X_cat= X_cat.replace(np.nan, '')
```

```python
X_dumm = pd.get_dummies(X_cat, drop_first=True) # You can put X, not separating variables is needed.
X_dumm
```

### Concat DataFrames

```python
transformer= MinMaxScaler().fit(X_num)
X_minmax=transformer.transform(X_num)
X_minmax.shape
```

```python
X_minmax
```

```python
X_num_norm=pd.DataFrame(X_minmax, columns=X_num.columns)
X_num_norm
```

```python
X_num = X.select_dtypes(include = np.number)
X = pd.concat([X_num_norm, X_dumm], axis=1)
X.head()
```

### Linear Regression

Train-test split.
Apply linear regression.

```python
X_train, X_test, y_train, y_test = train_test_split(X_dumm, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
predictions  = model.predict(X_test)
```
